```python
import numpy as np
import pandas as pd

from LingerRegression import LingerRegressions
```

Här i "data" behöver du lägga in din väg till filen du ska använda dig av.

```python
data = pd.read_csv("housing.csv")
data.head()
```

```python
features = [
    'housing_median_age',
    'median_income',
    'total_rooms',
    'total_bedrooms',
    'population',
    'households'
]
X = data[features]
Y = data["median_house_value"]
```

```python
X = X.fillna(X.mean())
Y = Y.fillna(Y.mean())
```

```python
data['median_house_value'].value_counts()
```

```python
data['ocean_proximity'].value_counts()
```

```python
model = LingerRegressions()

model.fit(X, Y)

print("beta-värden:", model.b)
print("prediktioner:", model.predict(X))
print("SSE:", model.SSE(X, Y))
print("varians:", model.regression_variance(Y))
print("RMSE:", model.rmse(X, Y))
```

```python
np.random.seed(42)

n = len(X)
indices = np.random.permutation(n)

split = int(n * 0.8)

train_idx = indices[:split]
test_idx  = indices[split:]

X_train = X[train_idx]
X_test  = X[test_idx]

Y_train = Y[train_idx]
Y_test  = Y[test_idx]

print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
```

```python
model = LingerRegressions()
model.fit(X_train, Y_train)

print("Antal datapunkter:", model.n)
print("Antal features:", model.d)
print("Beta-vektor:")
print(model.b)
```

```python
preds = model.predict(X_test)

print("Första 5 prediktioner:")
print(preds[:5])
```

```python
print("SSE:", model.SSE(X_test, Y_test))
print("RMSE:", model.rmse(X_test, Y_test))
print("Std(Y_test):", model.standard_deviation(Y_test))
```

```python
rmse = model.rmse(X_test, Y_test)
std = model.standard_deviation(Y_test)

print("RMSE / Std =", rmse / std)
```
