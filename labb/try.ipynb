```python
from LingerRegression import LingerRegressions
import pandas as pd

df = pd.read_csv("housing.csv")
```

```python
features = [
    'housing_median_age',
    'median_income',
    'total_rooms',
    'total_bedrooms',
    'population',
    'households'
]

X = df[features].dropna()
Y = df["median_house_value"].loc[X.index]
```

```python
model = LingerRegressions()
model.fit(X, Y)
```

```python
Y_hat = model.predict(X)
```

```python
print("Beta:", model.b)
print("n:", model.n)
print("d:", model.d)

print("SSE:", model.SSE(X, Y))
print("RMSE:", model.rmse(X, Y))
```
